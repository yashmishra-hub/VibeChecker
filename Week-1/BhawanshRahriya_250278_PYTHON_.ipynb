***Functions***

(Tip: Before starting this section, brush up on strings in python)

Write a function defintion that:
1. Adds the string "Hello, " to the beginning and "!" to the end of any given input.
2. Takes in two numbers, squares each number, then returns the sum of both squares.
3. Takes in value and returns the count of the nubmer of vowels in a sequence.
4. Takes in string and returns the string without any vowels.
5. Takes in a list of strings and returns the shortest string in the list.
6. Takes in sequence of numbers and returns the odd numbers in a list.
7. Takes in sequence of numbers and returns all the numbers added together.

```python
# Adds the string "Hello, " to the beginning and "!" to the end of any given input.

def helolfunc(x):
   x=f"Hello {x} !"
   return x;
# Takes in two numbers, squares each number, then returns the sum of both squares.

def sumsq( x, y):
  return x**2+y**2;
# Takes in value and returns the count of the nubmer of vowels in a sequence.

def countvowels( x):
  return x.count("a")+x.count("e")+x.count("i")+x.count("o")+x.count("u")
# Takes in string and returns the string without any vowels.

def novow( x):
  return x.replace("a","").replace("e","").replace("i","").replace("o","").replace("u","");
# Takes in a list of strings and returns the shortest string in the list.

def shortstring(x):
  minstring=x[0]
  for y in range(len(x)):
    if len(x[y])<len(minstring):
      minstring=x[y]
  return minstring;
# Takes in sequence of numbers and returns the odd numbers in a list.

def oddones(x):
  list(x)
  y=[z for z in x if z%2!=0 ]
  return y
# Takes in sequence of numbers and returns all the numbers added together.

def addall(x):
  list(x)
  y=0
  for i in x:
    y+=i
  return y

# usage
print(sumsq(2,4))
print(countvowels(" a baba black sheep"))
print(novow(" a baba black sheep"))
print(shortstring(["cow","buffalo","e","black"]))
print(oddones([1,2,3,4,56,2,34]))
print(addall([1,2,3,4,56,2,34]))
```
