Write a function defintion that:

```python
# answer for 1 and 2
def append(x):
    return("Hello"+x+"!")
def sum_sq(x,y):
    return(x*x + y*y)
y = append("World")
print(y)
print(sum_sq(2,3))

```

```python
#answer for 3
def vowel_counter(x):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in x:
      if ch in vowels:
        count+=1
    return count
x = "hello world"
print(vowel_counter(x))
```

```python
#answer for 4
def vowel_remover(x):
  vowels = "aeiouAEIOU"
  i = 0
  for ch in x:
    if ch in vowels:
      x = x[:i] + x[(i+1):]
    else:
      i += 1
  return(x)
vowel_remover("MokshBhatt")
```

```python
#answer for 5
def shortest_str(lst):
  shortest = lst[0]
  for s in lst:
    if len(s)<len(shortest):
      shortest = s
  return shortest
words = ["Moksh", "IITK", "Kanpur", "python"]
shortest_str(words)
```

```python
#answer for 6 & 7
def odds(lst):
  odd_lst = []
  for num in lst:
    if num%2 == 1:
      odd_lst.append(num)
  return odd_lst
def lst_sum(lst):
  sum = 0
  for num in lst:
    sum += num
  return sum
lst = [2,4,8,1,5,0]
print(odds(lst))
print(lst_sum(lst))
```

***Sets***

(Read up on sets in python)

Write a function definition that:
1. Takes in a list and returns a set with only the unique values from that list.
2. Takes two lists and returns a single set with the values that each list has in common.

(Bonus - Try doing the above questions from scratch using loops)

```python
def only_unique(lst):
  return set(lst)
def common(lst1,lst2):
  return set(lst1).intersection(set(lst2))
lst1 = [1,2,3,4,5,6]
lst2 = [4,5,6,7,8,9]
print(only_unique(lst1))
print(common(lst1,lst2))
```
