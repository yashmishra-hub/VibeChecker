***Consider a list l = [1,12,43,4,15,26].***


1. Write this list in python code.
2. Append a number to this list.
3. Sort this list in increasing order.
4. Reverse the sorted list.
5. Repeat questions (1-4) using loops and list indexing (Tip: if you can't figure out sorting, don't worry, you may skip it for now, those interested can look up "bubble sort" to get a hint).

```python
# your answer here
l = [1, 12, 43, 4, 15, 26]
print("Original list:", l)

l.append(30)
print("After append:", l)

l.sort()
print("Sorted list:", l)

l.reverse()
print("Reversed list:", l)

l = l + [30]
print("After append:", l)

n = len(l)
for i in range(n):
    for j in range(0, n-i-1):
        if l[j] > l[j+1]:
            l[j], l[j+1] = l[j+1], l[j]
print("Sorted list:", l)

reversed_list = []
for i in range(len(l)-1, -1, -1):  # loop backwards
    reversed_list.append(l[i])
print("Reversed list:", reversed_list)


```

Write a function defintion that:
1. Adds the string "Hello, " to the beginning and "!" to the end of any given input.
2. Takes in two numbers, squares each number, then returns the sum of both squares.
3. Takes in value and returns the count of the nubmer of vowels in a sequence.
4. Takes in string and returns the string without any vowels.
5. Takes in a list of strings and returns the shortest string in the list.
6. Takes in sequence of numbers and returns the odd numbers in a list.
7. Takes in sequence of numbers and returns all the numbers added together.

```python
# your answer here
def greet(name):
    return "Hello, " + name + "!"
def sum_of_squares(a, b):
    return a**2 + b**2
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    result = ""
    for char in s:
        if char not in vowels:
            result += char
    return result
def shortest_string(strings):
    return min(strings, key=len)
def odd_numbers(seq):
    result = []
    for num in seq:
        if num % 2 != 0:
            result.append(num)
    return result
def sum_all(seq):
    total = 0
    for num in seq:
        total += num
    return total
```

Write a function definition that:
1. Takes in a list and returns a set with only the unique values from that list.
2. Takes two lists and returns a single set with the values that each list has in common.

(Bonus - Try doing the above questions from scratch using loops)

```python
def unique_values(lst):
    return set(lst)
print(unique_values([1, 2, 2, 3, 4, 4, 5]))
def common_values(lst1, lst2):
    return set(lst1) & set(lst2)   # intersection
print(common_values([1, 2, 3, 4], [3, 4, 5, 6]))

def unique_values_loop(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
print(unique_values_loop([1, 2, 2, 3, 4, 4, 5]))

def common_values_loop(lst1, lst2):
    common_list = []
    for item in lst1:
        if item in lst2 and item not in common_list:
            common_list.append(item)
    return common_list
print(common_values_loop([1, 2, 3, 4], [3, 4, 5, 6]))

```

**Perform the following operations on the csv file:**

1. Read the csv into a pandas dataframe and back it up (you will thank yourself for doing this when you mess up something while manipulating the dataframe).
2. Explore the dataframe (find out the different ways you can do this).
3. Play around with each column of the dataframe and figure out ways to modify them according to your needs.
4. Read up on ``iloc`` and demonstrate it on the dataframe.
5. Handle ``null`` values in the dataframe.
6. Try dropping rows/columns.
7. Filter rows where a string column starts with a specific letter.
8. Create a new column as a combination of two existing columns (this is feature engineering which you may use in your ML journey).

```python
import pandas as pd
import numpy as np


url = "https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv"
df = pd.read_csv(url)
df_backup = df.copy()
print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(df.head())


print("\n=== DATAFRAME EXPLORATION ===")
print("1. Basic info:")
print(df.info())
print("\n2. First 5 rows:")
print(df.head())
print("\n3. Describe numerical columns:")
print(df.describe())
print("\n4. Columns:")
print(df.columns.tolist())
print("\n5. Data types:")
print(df.dtypes)
print("\n6. Missing values:")
print(df.isnull().sum())


print("\n=== COLUMN MANIPULATION ===")

df['date_clean'] = pd.to_datetime(df['date'])
df['price_per_sqft'] = df['price'] / df['sqft_living']
df['bedrooms_category'] = pd.cut(df['bedrooms'], bins=[0, 2, 4, 6, 10], labels=['Small', 'Medium', 'Large', 'XL'])  # Categorize
df['price_log'] = np.log(df['price'] + 1)  # Log transformation for skewed data
print("Sample modified columns:")
print(df[['price_per_sqft', 'bedrooms_category', 'price_log']].head())


print("\n=== ILOC DEMONSTRATION ===")
print("First 3 rows, first 3 columns:")
print(df.iloc[:3, :3])
print("\nRow 100, columns 0, 2, 3:")
print(df.iloc[100, [0, 2, 3]])
print("\nRows 50-55, columns 1-4:")
print(df.iloc[50:55, 1:4])


print("\n=== NULL VALUE HANDLING ===")
print("Null values before handling:")
print(df.isnull().sum())

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown', inplace=True)
print("\nNull values after handling:")
print(df.isnull().sum())


print("\n=== DROPPING ROWS/COLUMNS ===")
print(f"Shape before dropping: {df.shape}")
df_dropped = df.drop(columns=['date_clean', 'price_per_sqft']).copy()  # Drop some columns
df_dropped = df_dropped.dropna(subset=['price'])  # Drop rows with null price
print(f"Shape after dropping: {df_dropped.shape}")


print("\n=== STRING FILTERING ===")

filtered_zip = df[df['zipcode'].astype(str).str.startswith('98')]
print(f"Houses with zipcode starting with '98': {len(filtered_zip)}")
print(filtered_zip[['id', 'zipcode', 'price']].head())


print("\n=== FEATURE ENGINEERING ===")
df['price_per_bedroom'] = df['price'] / (df['bedrooms'] + 1)  # +1 to avoid division by zero
df['living_space_ratio'] = df['sqft_living'] / (df['sqft_lot'] + 1)
print("New engineered features:")
print(df[['price_per_bedroom', 'living_space_ratio']].head())
print("\nCorrelation with price:")
print(df[['price', 'price_per_bedroom', 'living_space_ratio']].corr())

print("\n=== BACKUP RESTORATION ===")
df = df_backup.copy()
print("Restored from backup!")

```
