***Consider a list l = [1,12,43,4,15,26].***


1. Write this list in python code.
2. Append a number to this list.
3. Sort this list in increasing order.
4. Reverse the sorted list.
5. Repeat questions (1-4) using loops and list indexing (Tip: if you can't figure out sorting, don't worry, you may skip it for now, those interested can look up "bubble sort" to get a hint).

```python
L = [1, 12, 43, 4, 15, 26]
print(f"Original List: {L}")

L.append(10)
print(f"After Append (10): {L}")

n = len(L)

for i in range(n - 1):
    for j in range(n - 1 - i):
        if L[j] > L[j + 1]:
            L[j], L[j + 1] = L[j + 1], L[j]

print(f"After Bubble Sort (Sorted): {L}")

L.reverse()
print(f"After Reverse: {L}")
```

***Functions***

(Tip: Before starting this section, brush up on strings in python)

Write a function defintion that:
1. Adds the string "Hello, " to the beginning and "!" to the end of any given input.
2. Takes in two numbers, squares each number, then returns the sum of both squares.
3. Takes in value and returns the count of the nubmer of vowels in a sequence.
4. Takes in string and returns the string without any vowels.
5. Takes in a list of strings and returns the shortest string in the list.
6. Takes in sequence of numbers and returns the odd numbers in a list.
7. Takes in sequence of numbers and returns all the numbers added together.

```python
def greet_string(s):
    return "Hello, " + s + "!"

def sum_of_squares(num1, num2):
    return (num1 *num1) + (num2 *num2)

def count_vowels(sequence):
    vowels = "aeiou"
    count = 0
    for char in sequence.lower():
        if char in vowels:
            count += 1
    return count

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    new_string = ""
    for char in s:
        if char not in vowels:
            new_string += char
    return new_string

def find_shortest_string(str_list):
    if not str_list:
        return ""

    shortest = str_list[0]
    for s in str_list:
        if len(s) < len(shortest):
            shortest = s
    return shortest

def get_odd_numbers(num_sequence):
    odd_list = []
    for num in num_sequence:
        if num % 2 != 0:
            odd_list.append(num)
    return odd_list

def sum_all_numbers(num_sequence):
    total = 0
    for num in num_sequence:
        total += num
    return total

print(f"1.greet_string('world')->{ greet_string('world')}")
print(f"2. sum_of_squares(3, 4) -> {sum_of_squares(3, 4)}")
print(f"3. count_vowels('Programming') -> {count_vowels('Programming')}")
print(f"4. remove_vowels('Elephant') -> {remove_vowels('Elephant')}")
print(f"5. find_shortest_string(['a', 'long', 'cat']) -> {find_shortest_string(['a', 'long', 'cat'])}")
print(f"6. get_odd_numbers([1, 2, 3, 4, 5]) -> {get_odd_numbers([1, 2, 3, 4, 5])}")
print(f"7. sum_all_numbers([10, 20, 30]) -> {sum_all_numbers([10, 20, 30])}")
```

3. Create a 3x3 matrix with values ranging from 0 to 8
4. Use NumPy to generate a random number between 0 and 1
5. Create an array of 20 linearly spaced points between 0 and 1

```python
import numpy as np

print("## NUMPY SECTION")

array_of_zeros = np.zeros(10)
print(f"1. Array of 10 zeros:\n{array_of_zeros}")

array_of_fives = np.full(10, 5)
print(f"2. Array of 10 fives:\n{array_of_fives}")

matrix_3x3 = np.arange(9).reshape(3, 3)
print(f"3. 3x3 Matrix (0-8):\n{matrix_3x3}")

random_float = np.random.rand()
print(f"4. Random float (0-1): {random_float}")

linear_array = np.linspace(0, 1, 20)
print(f"5. 20 linearly spaced points:\n{linear_array}")

print("\n## FUNCTIONS SECTION")

def greet_string(s):
    return "Hello, " + s + "!"

def sum_of_squares(num1, num2):
    return (num1 ** 2) + (num2 ** 2)

def count_vowels(sequence):
    vowels = "aeiou"
    count = 0
    for char in sequence.lower():
        if char in vowels:
            count += 1
    return count

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join([char for char in s if char not in vowels])

def find_shortest_string(str_list):
    if not str_list:
        return ""
    return min(str_list, key=len)

def get_odd_numbers(num_sequence):
    return [num for num in num_sequence if num % 2 != 0]

def sum_all_numbers(num_sequence):
    return sum(num_sequence)

print(f"1. greet_string('World'): {greet_string('World')}")
print(f"7. sum_all_numbers([1, 2, 3]): {sum_all_numbers([1, 2, 3])}")
```

**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

2. Scatter Plot
3. Boxplot

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)

```python
import matplotlib.pyplot as plt
import numpy as np
import random
hist_data = np.random.normal(loc=100, scale=15, size=1000)

x_scatter = np.linspace(0, 10, 50)
y_scatter = 2 * x_scatter + np.random.randn(50) * 3

group1 = np.random.normal(60, 10, 20)
group2 = np.random.normal(70, 5, 20)
group3 = np.random.normal(85, 15, 20)
group4 = np.random.normal(100, 8, 20)
boxplot_data = [group1, group2, group3, group4]
boxplot_labels = ['Group A', 'Group B', 'Group C', 'Group D']

pie_data = [30, 25, 15, 30]
pie_labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
pie_colors = ['gold', 'yellowgreen', 'coral', 'lightskyblue']
pie_explode = (0.1, 0, 0, 0)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

ax1, ax2, ax3, ax4 = axes.flatten()

ax1.hist(hist_data, bins=30, color='skyblue', edgecolor='black')
ax1.set_title('1. Histogram of Sample Data')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')

ax2.scatter(x_scatter, y_scatter, color='darkgreen', marker='o')
ax2.set_title('2. Scatter Plot')
ax2.set_xlabel('Independent Variable (X)')
ax2.set_ylabel('Dependent Variable (Y)')

ax3.boxplot(boxplot_data, labels=boxplot_labels, patch_artist=True)
ax3.set_title('3. Box Plot of Groups')
ax3.set_ylabel('Scores')

ax4.pie(pie_data, explode=pie_explode, labels=pie_labels, colors=pie_colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
ax4.axis('equal')
ax4.set_title('4. Pie Chart of Fruit Distribution')

plt.tight_layout(pad=3.0)
plt.show()
```
